# B.

**B1.** 전국 문헌정보학과 취업률 현황.csv 를 불러와 df라는 변수에 저장하시오. (3점) 

```python
df_b = pd.read_csv('전국 문헌정보학과 취업률 현황.csv', encoding = 'euc-kr')
df_b.head()
```

**B2.** 취업률이 높은 순서대로 정렬한 후, 취업률 상위 10개 학교명을 출력하시오. (5점)

```python
df_b.sort_values(by=['취업률'], ascending=False).head(10)
```

**B3.** 취업률이 평균이상인 학교들의 졸업자수 총합을 구하시오. (7점)

```python
#취업률 평균 
mean = df_b['취업률'].mean()
mean
```

```python
df_b[df_b['취업률']>=mean].sum()['졸업자수']
```

**B4.** 반복문과 조건문을 활용하여 학제를 숫자로(대학교 : 1, 일반대학원 : 2, 전문대학 : 3) 바꾸세요. (15점)

```python
for i in range(len(df_b)):
    if df_b.iloc[i, 0] == '대학교':
        df_b.iloc[i, 0] = 1
    elif df_b.iloc[i, 0] == '일반대학원':
        df_b.iloc[i, 0] = 2
    elif df_b.iloc[i, 0] == '전문대학':
        df_b.iloc[i, 0] = 3

df_b  
```

**B5.** 반복문과 조건문을 사용하지 말고 '주야구분' 열의 값들을 다음과 같이 변경하세요. (15점)
 
'일반과정_주간' -> '주간'  
'일반과정_야간' -> '야간'  
'학사학위전공심화(경력없음)_야간' -> '야간심화'  

```python
df_b['주야구분'] = df_b['주야구분'].replace('일반과정_주간', '주간')
df_b['주야구분'] = df_b['주야구분'].replace('일반과정_야간', '야간')
df_b['주야구분'] = df_b['주야구분'].replace('학사학위전공심화(경력없음)_야간', '야간심화')
df_b
```

# C.

**C1.** 출력값이 아래와 동일하도록 DataFrame을 생성하고 df라는 변수에 저장하시오. (3점)

|  | 이름	| 생년월일 | 
| ----- | ------------- | ------ |
| 0 | 남주혁 |	19940222 |
| 1 |	안효섭 | 19950417 |
| 2 | 김우빈  | 19890726|
|3 | 공유 |  19790710 |
|4 | 이준호 | 19900125|

```python
# code here
actor = {
    '이름' : ['남주혁', '안효섭', '김우빈', '공유', '이준호'],
    '생년월일' : ['19940222', '19950417', '19890726', '19790710', '19900125']
}
df_c = pd.DataFrame(actor)
df_c
```

**C2.** '생년월일' 열의 속성을 datetime으로 변경해주세요. (7점)

```python
df_c['생년월일'] = pd.to_datetime(df_c['생년월일'], errors='ignore')
```

```python
df_c.info()
```

**C3.** '생년월일' 열을 활용하여 '연도' , '월' , '일' , '나이' 라는 열을 생성해주세요. (8점)

```python
df_c['연도'] = df_c['생년월일'].dt.year
df_c['월'] = df_c['생년월일'].dt.month
df_c['일'] = df_c['생년월일'].dt.day
df_c['나이'] = pd.datetime.now().year - df_c['연도']
df_c
```

**C4.** 생일이 짝수일이고 나이가 30세 이상인 사람의 이름을 출력하시오. (7점)

```python
df_c.loc[((df_c['일']%2)==0)&(df_c['나이']>=30) , '이름']
```
