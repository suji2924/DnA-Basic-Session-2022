### Q1. 내장을 제외한 전복의 무게를 구하는 함수를 정의하고, 
### 함수를 적용해서 '전복-내장'이라는 컬럼을 abalone에 추가하세요 (11점)

```python
# 함수를 정의하세요 (6점)

def weight(x):
    w = x['whole_weight'] - x['viscera_weight']
    return w
```

```python
# 데이터에 함수를 적용하고, '전복-내장' 컬럼을 추가하세요 (5점)

abalone['전복-내장'] = weight(abalone)
abalone.head()
```

### Q2 . apply에서 lambda를 이용해서 'whole_weight'의 평균을 넘는 지에 대한 새로운 컬럼('heavy_w')를 정의하고, groupby를 이용하여 'sex'별 'heavy_w'가 평균을 넘는 개수가 몇개인지 구하세요 (11점)

```python
# apply에서 lambda를 이용해서 'whole_weight'의 평균을 넘는 지에 대한 새로운 컬럼('heavy_w')를 정의하세요 (6점)

abalone['heavy_w'] = abalone.apply(lambda x : x['whole_weight'] > abalone['whole_weight'].mean(), axis = 1)
abalone['heavy_w']
```

```python
abalone.head()
```

```python
# groupby를 이용하여 'sex'별 'heavy_w'가 평균을 넘는 개수가 몇개인지 구하세요 (5점)

abalone.groupby('sex')['heavy_w'].sum()
```

### Q3. sex별로 height,whole_weight들의 평균값을 구하시오. (6점)

agg를 사용하세요

```python
# 코드를 작성하세요

abalone.groupby('sex')[['height', 'whole_weight']].agg('mean')
```

### Q5. sex를 제외한 column 중 원하는 두 column을 정하고, 두 컬럼의 합을 구하는 groupby를 만들기  (12점)
집계함수, apply, agg, lambda 등 사용하여 두개를 작성하세요  
(컬럼과 함수는 다른것으로 사용하세요)  
(key값은 sex로)

```python
# (1) (6점)
# 선택 컬럼을 적어주세요 :  shucked_weight, viscera_weight

# 코드를 작성하세요

abalone.groupby('sex')[['shucked_weight', 'viscera_weight']].agg(sum)
```

### Q6. key값을 sex로 하고, 나머지 column들에 min, mean, max를 적용하고, 이름을 최소, 평균, 최대 로 바꾸기 (8점)

```python
# 코드를 작성하세요

abalone.groupby('sex').agg([('최소', 'min'), ('평균', 'mean'), ('최대', 'max')])
```

### Q7. pivot_table을 이용하여 abalone 데이터 살펴보기  (10점)
pivot_table을 사용하여 두가지를 작성해보세요  

```python
# (1) 코드를 작성하세요 (5점)

pd.pivot_table(abalone,
               values='whole_weight', 
               index='sex', 
               columns='rings', 
               aggfunc='mean', 
               fill_value=None, 
               margins=False, 
               dropna=True, 
               margins_name='All')
```

```python
# (2) 코드를 작성하세요 (5점)

abalone.pivot_table(values='diameter', 
               index='length', 
               columns='sex', 
               aggfunc='sum', 
               fill_value=None, 
               margins=True, 
               dropna=True, 
               margins_name='All sum')
```

#### 9-1. pivot_table을 이용하여 작성하세요 (6점)

#### 9-2. crosstab을 이용하여 작성하세요 (6점)
(crosstab의 경우 뒤에 소수점이 있을 수 있습니다)  

```python
# crosstab 사용한 코드를 작성해주세요

pd.crosstab(values=df_phone['duration'],
            index=[df_phone.month, df_phone.item],
            columns=df_phone['network'],
            aggfunc='sum').fillna(0)
```

### Q10. crosstab을 사용해서 2015-02에 사용수가 많은 item을 찾고,  2014.11~2015.03의 sms의 총 빈도수를 찾아보세요  (11점)

**hint: margins**

```python
# 코드를 작성해주세요 (10점)

pd.crosstab(values=df_phone['duration'],
            index=df_phone['month'],
            columns=df_phone['item'],
            aggfunc='count',
            margins=True).fillna(0)
```
