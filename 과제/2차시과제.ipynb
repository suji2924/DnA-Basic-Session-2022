### Q10. 아래 그림을 참고하여 (1)과 (2) 두 수를 사용하여 (3), (4), (5), (6)의 값을 출력하시오. [10점]

(총 4개의 값이 출력되어 있어야 함)

![formula](formula.png)

```python
a=472
b=385

aa=str(a)
bb=str(b)


answer3 = 100*(int(bb[-1])*int(aa[0])) + 10*(int(bb[-1])*int(aa[1])) + int(bb[-1])*int(aa[2])
print(answer3)
```

```python
answer4 = 100*(int(bb[1])*int(aa[0])) + 10*(int(bb[1])*int(aa[1])) + int(bb[1])*int(aa[2])
print(answer4)
```

```python
answer5 = 100*(int(bb[0])*int(aa[0])) + 10*(int(bb[0])*int(aa[1])) + int(bb[0])*int(aa[2])
print(answer5)
```

```python
answer6 = answer3 + (10*answer4) + (100*answer5)
print(answer6)
```
