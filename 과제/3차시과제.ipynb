### Q4. 이름과 년생이 dictionary에 담겨있을 때 번호를 부여하고 다음과 같이 출력하세요

출력결과
```
번호 : 1, 유아 : 96
번호 : 2, 지호 : 97
번호 : 3, 미미 : 96
번호 : 4, 아린 : 99
번호 : 5, 승희 : 96
번호 : 6, 효정 : 95
번호 : 7, 유빈 : 97
```

````python
info = {"유아" : 95, "지호" : 97, "미미" : 95, "아린" : 99, "승희" : 96, "효정" : 94, "유빈" : 97}

#dic 값 수정
for i, v in info.items():
    if i == '유아':
        info[i] = 96
    elif i == '미미':
        info[i] = 96
    elif i == '효정':
        info[i] = 95

#출력
for j, k, n in zip(range(1,8), info.keys(), info.values()):
    print(f'번호 : {j}, {k} : {n}')
````

### Q5. 숫자를 입력받아 주어진 수가 1이 될때까지 다음 작업을 반복하고 반복한 횟수를 구하시오

(단, 작업을 500번 반복해도 1이 되지 않는다면 –1을 출력하시오.)
```
1-1. 입력된 수가 짝수라면 2로 나눈다.
1-2. 입력된 수가 홀수라면 3을 곱하고 1을 더한다.
2. 결과로 나온 수에 같은 작업을 1이 될 때까지 반복한다.

[ 입력예시 ] 6  
[ 출력예시 ] 8 (6 -> 3 -> 10 -> 5 -> 16 -> 8 -> 4 -> 2 -> 1)
```

````python
num = int(input('숫자를 입력하시오 : '))
t = 0

while num!=1:
    
    if t>=500:
        print(-1)
        break
    
    if (num%2)==0:
        num = num/2
        t+=1
    else:
        num = (num*3)+1
        t+=1

print(t)
````

### Q9. 오바마 연설문을 가져와 가장 많이 언급되는 단어와 그것의 횟수를 출력하세요
       (1) words 변수에 담긴 문자열을 띄어쓰기 기준으로 나눈다.
       (2) words의 각 단어들을 key로, 각 단어의 횟수를 value로 하는 dictionary를 만든다.
       (3) value가 가장 큰 key와 value를 출력한다!

````python
# 텍스트 화일을 열고 words라는 변수에 저장
with open('Obama_words.txt', 'r') as f:
    words = f.read()
    f.close()
    
print(words)
````

````python
# 이 셀에 문제를 푸세요!

#띄어쓰기 기준 나누기
s = words.split()
print(s)
````

````python
#dic 만들기
dic={}
for i in s:
    if i not in dic:
        c = s.count(i)
        dic[i] = c
dic
````

````python
#value값 가장 큰 key, value 출력

{k : v for k,v in dic.items() if max(dic.values()) == v}
````
